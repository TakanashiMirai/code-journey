# Python 列表
```python
shopping_list = ["键盘", "键帽"]

# 方法 => 对象.方法名()   函数 => 函数名(操作对象)

#列表是可变的，其他不可变
# 增加列表元素
shopping_list.append("显示器")
# 删除列表元素
shopping_list.remove("显示器")
# 求长度
len(shopping_list)
# 索引
print(shopping_list[1]) # 从零开始数
shopping_list[1] = "鼠标" # 直接索引赋值
# max min sort函数都可使用
```

# Python 字典
字典用于存储键和值，键用来查找值
```python
contact = {
    "A" = "1234567"
    "B" = "7654321"
}

# 获取键的值
contact["A"]

# 元组 tuple 元组可以放多个元素
name_tuple = ("A", "32") # 不可变

contact = {
    ("A", "32") = "1234567"
}

# 添加值 如果键已存在，则元素会覆盖
contact["C"] = "999"
# 删除值 如果键不存在，则会报错
del contact["A"]
# 字典长度
len(contact)
```
